```python
# cleanning database. The n/a** in the columns of debt means that are less than 50 citizens in the county.
debt_clean_df=debt_clean_df1.replace(['n/a**'], 49)
debt_clean_df.head()
```

### Pivot_table()

The pivot operation allows us to build a new dataframe from another, so that the values of the column 'party' become column names of the new dataframe.

A pivot table is a summary table (data arranged in rows and columns) that groups data from another table or larger database. It is used as a tool for processing large amounts of data.

```python
# new dataframe using pivot_table for county_fips
df=elections_df1.pivot_table(index='county_fips', columns = 'party', values = 'candidatevotes', aggfunc='sum')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df2=elections_df1.pivot_table(index='county_name', columns = 'party', values = 'candidatevotes', aggfunc='sum')
df2.head()
```

```python
# filling NaN values because they mean that the have zero votes, so is important to keep
df2.fillna(0, inplace=True)
df2.head()
```

```python
# verifying data types for building tables in sql
df2.dtypes
```

```python
#Creating the winner column in order to know which party won
df2['WINNER'] = np.where((df2['DEMOCRAT'] > df2['REPUBLICAN']),1 , 0)
```

```python
df2.count()
```

```python
df2.head()
```

```python
# Transform clean data to csv
df2.to_csv("elections2020_clean.csv")
```

```python
# #Connecting clean data to postgresql
df2.to_sql('elections2020_clean', con = engine, if_exists = 'replace')
```

```python
elections_df = county_elections_df.drop(columns = ['office', 'candidate', 'totalvotes', 'version', 'mode'])
elections_df
```

```python
population_df['county_name'] = population_df['CTYNAME'].str.split(' ').str[0]
population_df['county_name'] = population_df['county_name'].str.upper()
```

```python
poll_pop_df = pd.merge(elections_df,population_df, how = 'left', on =['county_name'])
poll_pop_df
```

```python
final_df = pd.merge(poll_pop_df, debt_df, how='left', on = ['county_name'])
final_df
```

```python
final_df[['Share with Open Auto/Retail Debt, All', 'Share with Open Auto Debt, Communities of Color', 'Share with Open Auto Debt, Majority White Communities']] = final_df[['Share with Open Auto/Retail Debt, All', 'Share with Open Auto Debt, Communities of Color', 'Share with Open Auto Debt, Majority White Communities']].replace(to_replace = ['n/a*', 'n/a**'], value=0).astype(float)

final_df.dtypes
```

```python
final_df['Share with Open Auto/Retail Debt, All'] = final_df['Share with Open Auto/Retail Debt, All']*final_df['POPESTIMATE2019']
```

```python
final_df['Share with Open Auto Debt, Communities of Color'] = final_df['Share with Open Auto Debt, Communities of Color']*final_df['POPESTIMATE2019']
```

```python
final_df['Share with Open Auto Debt, Majority White Communities'] = final_df['Share with Open Auto Debt, Majority White Communities']*final_df['POPESTIMATE2019']
```

```python
final1_df = final_df.drop(columns=['CTYNAME','county','NAME'])
```

```python
final1_df[['Share with Open Auto/Retail Debt, All', 'Share with Open Auto Debt, Communities of Color', 'Share with Open Auto Debt, Majority White Communities']] = final1_df[['Share with Open Auto/Retail Debt, All', 'Share with Open Auto Debt, Communities of Color', 'Share with Open Auto Debt, Majority White Communities']].replace(to_replace = 0, value=50).astype(float)
```

```python
final1_df = final1_df[final1_df['year'] == 2020]
final1_df
```

```python
#Declare a new list to encounter the candidates

party_options = []

#Declare a dictionary in which we will have the votes for each candidate

party_votes = {}

state_options = []

#print each row in the csv file
for index, row in final1_df.iterrows():
    
    state_name = row['state']
    
    if state_name not in state_options:

        #add the total vote count

        total_votes = row['candidatevotes']

        #Find the candidates that ran in the election

        party_name = row['party']

        #Add it to the list of candidates
        if party_name not in party_options:
            #add candidate name to candidates list
            party_options.append(party_name)
            #Begin tracking that candidates vote count
            party_votes[party_name] = 0
            party_votes[party_name] = total_votes
            
        state_options.append(state_name)
            

```

```python
poll_pop_df = pd.merge(elections_df, population_df, how ='left', on ='county_name')
poll_pop_df.head(20)
```

```python
final_df = pd.merge(poll_pop_df1, debt_df, how='left', on = ['county_name'])
final_df
```

```python
final1=final_df.drop_duplicates(subset=['county_name','party'])
```

```python
final1['candidatevotes'].sum()
```

```python
final1[['Share with Open Auto/Retail Debt, All', 'Share with Open Auto Debt, Communities of Color', 'Share with Open Auto Debt, Majority White Communities']] = final_df[['Share with Open Auto/Retail Debt, All', 'Share with Open Auto Debt, Communities of Color', 'Share with Open Auto Debt, Majority White Communities']].replace(to_replace = ['n/a*', 'n/a**'], value=0).astype(float)

final1.dtypes
```

```python
final1['Share with Open Auto/Retail Debt, All'] = final_df['Share with Open Auto/Retail Debt, All']*final_df['POPESTIMATE2019']
```

```python
final1['Share with Open Auto Debt, Communities of Color'] = final_df['Share with Open Auto Debt, Communities of Color']*final_df['POPESTIMATE2019']
```

```python
final1['Share with Open Auto Debt, Majority White Communities'] = final_df['Share with Open Auto Debt, Majority White Communities']*final_df['POPESTIMATE2019']
```

```python
final1_df = final1.drop(columns=['CTYNAME','NAME'])
```

```python
final1_df[['Share with Open Auto/Retail Debt, All', 'Share with Open Auto Debt, Communities of Color', 'Share with Open Auto Debt, Majority White Communities']] = final1_df[['Share with Open Auto/Retail Debt, All', 'Share with Open Auto Debt, Communities of Color', 'Share with Open Auto Debt, Majority White Communities']].replace(to_replace = 0, value=50).astype(float)
```

```python
#Declare a new list to encounter the candidates



#Declare a dictionary in which we will have the votes for each candidate

party_votes = {}
county_options = []
county_winner = []

#print each row in the csv file
for index, row in final1_df.iterrows():
    
    party_options = []
    
    county_name = row['county_name']
    
    if county_name not in county_options:

        #add the total vote count

        total_votes = row['candidatevotes']

        #Find the votes that won in the counties

        party_name = row['party']

        #Add it to the list of candidates
        if party_name not in party_options:
            #add candidate name to candidates list
            party_options.append(party_name)
            #Begin tracking that candidates vote count
            party_votes[party_name] = 0
            party_votes[party_name] = total_votes
            
            
        county_options.append(county_name)
        
    county_win = max(party_votes)
        
    if county_win == "DEMOCRAT":
        
        winner = 1
    else: 
        
        winner = 0
            
    county_winner.append(winner)
    
        

```
